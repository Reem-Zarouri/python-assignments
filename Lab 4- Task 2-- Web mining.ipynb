```python
#Task 2
#Importing needed libraries
from bs4 import BeautifulSoup
import requests
import sys
```

```python
# desktop user-agent
#USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"

headers = {"user-agent": USER_AGENT}
r = requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250",headers=headers)

print("Status code: ", r.status_code)
```

```python
if r.status_code == 200:
    # resp = requests.get(URL, headers=headers)
    soup = BeautifulSoup(r.content, "html.parser")
else:
    print("Scrape Error")
```

```python
# The below code is only for reference
# finds each product from the store page
containers = soup.find("table", {"class": "chart full-width"})

container = containers.find_all("tr")

# name the output file to write to local disk
out_filename = "imbd22.csv"
# header of csv file to be written
headers = "Film,Rating \n"
```

```python
# opens file, and writes headers
out_filename = "imbd22.csv"
f = open(out_filename, "w+")
f.write(headers)

# loops over each product and grabs attributes about
# each product
for con in container:
    RR  = con.find_all('td', {"class": "ratingColumn imdbRating"})
    anchors = con.find_all('td', {"class": "titleColumn"})

    if anchors:

        title = con.find('td', {"class" : "titleColumn"})
        c = title.text

    if RR:
        rating = con.find("td", {"class" : "ratingColumn imdbRating"})
            
        b = rating.text
    
    print(c)
    print(b)
    print(c, b, file=open(out_filename, "a"))



f.close()  # Close the file
```
