```python
# as we know Clustering in data analysis refers to the process of grouping similar data points together based
#on their inherent characteristics or patterns.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df['Borough_Code'] = pd.factorize(df['Borough'])[0]
X = df[['Borough_Code', 'Cost']]
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
df['Cluster'] = kmeans.labels_
plt.scatter(df['Borough_Code'], df['Cost'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Borough Code')
plt.ylabel('Cost')
plt.title('Clustering of Borough and Cost')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt

# Performing one-hot encoding on the 'Borough' column    
one_hot = pd.get_dummies(df['Borough'])

# Combining the one-hot encoded features with the 'Cost' column
data = pd.concat([one_hot, df['Cost']], axis=1)

# Creating DBSCAN model
dbscan = DBSCAN(eps=1, min_samples=1)
clusters = dbscan.fit_predict(data)

# Scatter plot
plt.scatter(df['Borough'], df['Cost'], c=clusters, cmap='viridis')
plt.xlabel('Borough')
plt.ylabel('Cost')
plt.title('DBSCAN Clustering')
plt.show()

# Compute silhouette score
silhouette_avg = silhouette_score(data, clusters)
print("Silhouette Score:", silhouette_avg)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

borough = df['Borough']
cost = df['Cost']

# Filter the dataset to the first 100 records
borough = borough[:100]
cost = cost[:100]
# ChatGPT helped here
# because it kept throwing erorr.
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')  
borough_encoded = encoder.fit_transform(borough.values.reshape(-1, 1))

# Perform linear regression
regression_model = LinearRegression()
X = borough_encoded
y = cost.values
regression_model.fit(X, y)

# Make predictions
new_borough = ['MANHATTAN', 'BROOKLYN', 'QUEENS']  
new_borough_encoded = encoder.transform(pd.Series(new_borough).values.reshape(-1, 1))
predicted_cost = regression_model.predict(new_borough_encoded)

# Print the predictions
for borough, cost in zip(new_borough, predicted_cost):
    print(f'Predicted cost for {borough}: ${cost:.2f}')
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error


# Splitting the data into features (X) and target (y)
X = df[['Borough']]
y = df['Cost']

# One-hot encoding the categorical variable (Borough) 
# ChatGPT helped here as well.
ct = ColumnTransformer([('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X_encoded = ct.fit_transform(X)

# Splitting the data into training and test sets
X_train, X_test = X_encoded[:100], X_encoded[100:]
y_train, y_test = y[:100], y[100:]

# Initializing and training the kNN regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Making predictions on the test set
y_pred = knn.predict(X_test)

# Calculating the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

#  A larger MSE indicates that the data points are dispersed widely around its central moment (mean),
# Error in this case means the difference between the observed values and the predicted ones.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder   


features = df[['Borough', 'Cost']]
target = df['Board']

# ChatGPT helped here.
# Encode categorical features to numerical values
label_encoder = LabelEncoder()
features['Borough'] = label_encoder.fit_transform(features['Borough'])

# Split the dataset into training and test sets
# with a test size = 30%
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Build and train the kNN classifier
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred = knn.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = knn.score(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

```python
import pandas as pd
from scipy import stats

# using z_score to detect outlier in the column 'Cost'.
selected_column = df['Cost']

# ChatGPT
# Calculate z-scores
z_scores = stats.zscore(selected_column)

# Set a threshold for outlier detection (e.g., z-score > 3 or z-score < -3)
threshold = 3

# Identify outliers
outliers = selected_column[abs(z_scores) > threshold]

# Print the outliers
print(outliers)
```

```python
import pandas as pd

selected_column = df['Cost']

# I want to use another method to detect the ouliers in our dataset,
# and the second method is IQR(Interquantile Range).
# YouTube vedio gave me a better understanding about IQR.
# Calculate the interquartile range (IQR)
Q1 = selected_column.quantile(0.25)
Q3 = selected_column.quantile(0.75)
IQR = Q3 - Q1

# Set the threshold for outlier detection
threshold = 1.5

# Identify outliers
lower_bound = Q1 - threshold * IQR
upper_bound = Q3 + threshold * IQR
outliers = selected_column[(selected_column < lower_bound) | (selected_column > upper_bound)]

# Print the outliers
print(outliers)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Performing PCA component analysis
train_data = X_train.copy()
train_data['Board'] = y_train

# Select only the 'Borough', 'Cost', and 'Board' columns
train_data_subset = train_data[['Borough', 'Cost', 'Board']]

# Perform PCA on the selected features.
# Select the number of components to retain
# ChatGPT helped here.
pca = PCA(n_components=2)  
principal_components = pca.fit_transform(train_data_subset[['Borough', 'Cost']])
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Combine the principal components with the target variable
pca_data = pd.concat([principal_df, train_data_subset['Board']], axis=1)

# Plot the pairplot of the principal components with the target variable
sns.pairplot(pca_data, hue='Board')
accuracy = knn.score(X_test, y_test)
print('Accuracy:', accuracy)
# Display the plot
plt.show()
```
